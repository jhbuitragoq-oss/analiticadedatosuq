```python
# 1) POBLACION TOTAL = La población total corresponde al número total de registros en la base de datos.
df_cgsse['FEX_C18'].sum()
```

```python
# POBLACION EN EDAD DE TRABAJAR = Se filtran las personas con edad mayor o igual a 15 años, que corresponden a la población en edad de trabajar.
PET = df_cgsse[df_cgsse['P6040']>=15]['FEX_C18'].sum()
print(f'Poblacion en edad de trabajar: {PET}')
```

```python
# POBLACION QUE NO ESTA EN EDAD DE TRABAJAR = Se filtran las personas menores de 15 años y se suma el factor de expansión para estimar la población total representada.
PNET = df_cgsse[df_cgsse['P6040']<15]['FEX_C18'].sum()
print(f'Poblacion que no está en  edad de trabajar: {int(PNET)}')
```

```python
# POBLACION ECONOMICAMENTE ACTIVA E INACTIVA = La PEA incluye las personas que estan trajando y estan buscando trabajo y la PEI Son las personas que no trabajan ni buscan trabajo.
PEA = df_ft[df_ft['FT']==1]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa: {PEA}')

PEI = df_ft[df_ft['FFT']==1]['FEX_C18'].sum()
print(f'Poblacion Economicamente inactiva: {int(PEI)}')
```

```python
# TASA OCUPACION = La tasa de ocupación muestra qué porcentaje de la población en edad de trabajar está ocupada.
O = df_ocupados['FEX_C18'].sum()
print(f'Poblacion ocupada: {int(O)}')

PET = df_ft['FEX_C18'].sum()
print(f'Poblacion En edad de trabajar : {int(PET)}')

TO = O/PET
print(f'Tasa de Ocupacion: {TO:.2%}')
```

```python
# TASA DE DESEMPLO = La tasa de desempleo mide el porcentaje de la población activa que no tiene trabajo pero lo está buscando.
D = df_desocupados[df_desocupados['DSI']==1]['FEX_C18'].sum()
print(f'Poblacion Desempleada: {D}')

TD = D/PEA
print(f'Tasa de Desempleo: {TD*100}%')
```

```python
# 2)Calcule las tasas de desempleo para hombres y mujeres en Risaralda, Caldas , Tolima y Quindío. ¿Cuál es el departamento con mayor y menor tasa de desempleo de los 4 ?
df_d = pd.merge(df_desocupados, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']],
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')

df_ft = pd.merge(df_ft, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']],
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#QUINDIO
# 1. Calcular cantidad de hombres desempleados en el QUINDIO
HDQ = df_desocupados[(df_d['DSI']==1) & (df_d['DPTO']==63) & (df_d['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en el Quindio: {HDQ}')

# 2. Calcular cantidad de hombres en la PEA del QUINDIO
PEAHQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==63) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en el quindío: {PEAHQ}')

# 3. Calcular la tasa de desempleo en hombres del QUINDIO
TDHQ = HDQ/PEAHQ
print(f'Tasa de Desempleo en hombres del quindío: {TDHQ*100}%')
```

```python
# 1. Calcular cantidad de MUJERES desempleadas en el QUINDIO
MDQ = df_desocupados[(df_d['DSI']==1) & (df_d['DPTO']==63) & (df_d['P3271']==2)]['FEX_C18'].sum()
print(f'Mujeres Desempleadas en el Quindio: {MDQ}')

# 2. Calcular cantidad de mujeres en la PEA del QUINDIO
PEAMQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==63) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en el quindío: {PEAMQ}')

# 3. Calcular la tasa de desempleo en mujeres del QUINDIO
TDMQ = MDQ/PEAMQ
print(f'Tasa de Desempleo en mujeres del quindío: {TDMQ*100}%')
```

```python
# TASA DESEMPLEO QUINDIO
DTQ = HDQ + MDQ
PEATQ = PEAHQ + PEAMQ

TDQ = DTQ / PEATQ

print(f"Tasa de desempleo total Quindío: {TDQ*100:.2f}%")
```

```python
#RISARALDA
# Hombres desempleados en RISARALDA
HDR = df_desocupados[(df_d['DSI']==1) & (df_d['DPTO']==66) & (df_d['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleadas en Risaraldoa: {HDR}')

# Hombres en la PEA
PEAHR = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==1)]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de Hombres en Risaralda: {PEAHR}')

# Tasa desempleo hombres
TDHR = HDR/PEAHR
print(f'Tasa de desempleo hombres Risaralda: {TDHR*100}%')
```

```python
# 1. Calcular cantidad de MUJERES desempleadas en Risaralda
MDR = df_desocupados[(df_d['DSI']==1) & (df_d['DPTO']==66) & (df_d['P3271']==2)]['FEX_C18'].sum()
print(f'Mujeres Desempleadas en el Risaralda: {MDR}')

# 2. Calcular cantidad de mujeres en la PEA del Risaraldo
PEAMR = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en el Risaralda: {PEAMR}')

# 3. Calcular la tasa de desempleo en mujeres del Risaralda
TDMR = MDR/PEAMR
print(f'Tasa de Desempleo en mujeres del Risaralda: {TDMR*100}%')
```

```python
# TASA DESEMPLEO RISARALDA
DTR = HDR + MDR
PEATR = PEAHR + PEAMR

TDR = DTR / PEATR

print(f"Tasa de desempleo total Risaralda: {TDR*100:.2f}%")
```
